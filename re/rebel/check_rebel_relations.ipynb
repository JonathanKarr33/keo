### Check Rebel Relation Set

This notebook checks whether all relations generated by rebel belong to the list of relations in the rebel github at rebel/main/data/relations_count.tsv

```python
import pandas as pd
```

**Get set of relations listed on rebel github**

```python
!wget https://raw.githubusercontent.com/Babelscape/rebel/main/data/relations_count.tsv
!mv relations_count.tsv relations_count.csv
```

```python
relation_set_df = pd.read_csv("relations_count.csv", delimiter='\t')
relation_set_df
```

```python
relation_set = list(relation_set_df["country"])
relation_set.append('country')
```

**Get all relations found by Rebel in results data**

```python
results_df = pd.read_csv("../../data/results/rebel/rebel_results_spacy_pipe.csv")
relations_found = results_df["relation"].unique()
```

**Check if all relations found by Rebel belong to relation set**

```python
for relation in relations_found:
    if relation not in relation_set:
        print(f'relation "{relation}" was generated by Rebel, but is not part of orignal set')
```

```python
'different from' in relation_set
```

```python
'said to be the same as' in relation_set
```

**All relations found by Rebel belong to the relation set**

The discrepancies found are a matter of incorrect parsing, not new relations
